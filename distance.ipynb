```python

listakm = []
listadeceps =[]
listadeerros =[]

for cep in listaceps['CEP']:
    try:
        navegador.get('https://www.google.com.br/maps/dir/{}/SP+Equipamentos,+Av.+Robert+Kennedy,+675+-+Vila+Planalto,+São+Bernardo+do+Campo+-+SP,+09895-003/'.format(cep))
        km = navegador.find_element(By.XPATH,'//*[@id="section-directions-trip-0"]/div[1]/div[1]/div[1]/div[2]/div').text
        listakm.append(km)
        listadeceps.append(cep)
        print(cep)
    except:
        listadeerros.append(cep)
```
